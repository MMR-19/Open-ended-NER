```python
import numpy as np
import pandas as pd
```

Import np arrrays


```python
entity_dict_1 = {
    'CrossNER · AI': np.load('entity_length_crossner_ai.npy'),
    'CrossNER · Literature': np.load('entity_length_crossner_literature.npy'),
    'CrossNER · Music': np.load('entity_length_crossner_music.npy'),
    'CrossNER · Politics': np.load('entity_length_crossner_politics.npy'),
    'CrossNER · Science': np.load('entity_length_crossner_science.npy'),
    'MultiNERD · EN': np.load('entity_length_multinerd_en.npy'),
    'MultiNERD · PT': np.load('entity_length_multinerd_pt.npy'),
    # 'E-NER': np.load('entity_length_ener.npy'),
    # 'LeNER-Br': np.load('entity_length_lener.npy'),
    # 'NeuralShift': np.load('entity_length_neuralshift.npy')
}

entity_dict_2 = {
    'E-NER': np.load('entity_length_ener.npy'),
    'LeNER-Br': np.load('entity_length_lener.npy'),
    'NeuralShift': np.load('entity_length_neuralshift.npy')
}

entity_df_1 = pd.DataFrame(dict([(k, pd.Series(v)) for k, v in entity_dict_1.items()]))

entity_df_2 = pd.DataFrame(dict([(k, pd.Series(v)) for k, v in entity_dict_2.items()]))
```

```python
# horizontal violinplot of entity lengths from seaborn
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
# show only median inner
sns.violinplot(data=entity_df_1, orient='h', cut=0, inner=None)
medians = entity_df_1.median()
# overlay median markers (white circle with black edge)
for i, m in enumerate(medians.values):
    plt.scatter(m, i, facecolor='white', edgecolor='black', s=20, zorder=3, marker='o')

plt.xticks(range(1, 15))
#save as svg
plt.savefig('entity_length_violinplot_1.svg', format='svg', dpi=1200)
```

```python
# horizontal violinplot of entity lengths from seaborn
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 3*6/7))
# show only median inner
sns.violinplot(data=entity_df_2, orient='h', cut=0, inner=None)
medians = entity_df_2.median()
# overlay median markers (white circle with black edge)
for i, m in enumerate(medians.values):
    plt.scatter(m, i, facecolor='white', edgecolor='black', s=20, zorder=3, marker='o')
    print(f"{entity_df_2.columns[i]} median: {m}")

plt.xticks([1, 2, 3, 5, 10, 20, 30, 40, 50, 60])
#save as svg
plt.savefig('entity_length_violinplot_2.svg', format='svg', dpi=1200)
```
